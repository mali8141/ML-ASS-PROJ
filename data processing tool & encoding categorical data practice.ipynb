## encoding categorical data

```python
# creating enitial dataframe
vehicle =['car','bike','bike','car','car','bike','bike','car','car','bike']
vehicle_df= pd.DataFrame(vehicle, columns=['Vehicle'])
```

```python
vehicle_df
```

```python
#binary table
```

```python
vehicle_df['Vehicle'] = vehicle_df['Vehicle'].map({'car':1, 'bike':0})
```

```python
vehicle_df
```

```python
#multiple table
```

```python
ownership = ['new','old','emi','old','emi','emi','new','old','emi','new']
```

```python
vehicle_df['Ownership']=ownership
```

```python
vehicle_df
```

```python
df1=pd.get_dummies(vehicle_df['Ownership'])
```

```python
df1
```

```python
vehicle_df  = pd.concat([vehicle_df,df1],axis=1)
```

```python
vehicle_df
```

```python
vehicle_df.drop(['Ownership'],axis=1,inplace=True)
```
